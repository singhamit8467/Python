```python
# Importing Libraries and Loading File from Location
import pandas as pd
import matplotlib.pyplot as plt
x=pd.read_csv('Online Retail.csv')
```

```python
# Basic View of Retails Data
import seaborn as sns
x
```

```python
# Null value & Data Type Check
x.info()
```

```python
x['Quantity'].info()
```

```python
# Problem Areas Need further Investigation as Sales Quantity is -ve.
x.query('Quantity < 0')
```

```python
# Use Rank for Totalsales Decision Making
x['Ranking']=x.groupby(['StockCode'])['Totalsale'].rank()
```

```python
# Arranging Dataset with Descending Ranks
x.sort_values(['Ranking'],ascending=False)
```

```python
x['Ranking'].describe()
```

```python
# StockCode with Poor sales Numbers 
x.query('Ranking < 49')
```

```python
# Stock Code with Good sales Numbers 
x.query('Ranking > 49')
```

```python
# Weak TotalSale Countries Bottom(10)
x.groupby(x['Country']).sum().sort_values(['Totalsale'],ascending=True).drop(columns=['Ranking','CustomerID','UnitPrice']).head(10)
```

```python
# Countries with Strong Overall Totalsale Number's Top(10)
x.groupby(x['Country']).sum().sort_values(['Totalsale'],ascending=False).drop(columns=['Ranking','CustomerID','UnitPrice']).head(10)
```
